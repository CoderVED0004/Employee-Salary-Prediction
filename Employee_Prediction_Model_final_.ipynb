```python
#encoding is coverting all the categorical to numerical
data = pd.get_dummies(data, columns=['workclass'], drop_first=True)
```

```python
categorical_cols = ['occupation', 'education', 'marital-status',
                    'relationship', 'native-country', 'race']

data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
data['income'] = le.fit_transform(data['income'])
```

```python
x = data.drop(columns=['income'])  # Features (inputs)
y = data['income']
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Create an imputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Create a pipeline that first imputes and then fits the logistic regression model
model = Pipeline([
    ('imputer', imputer),
    ('logistic', LogisticRegression(max_iter=1000))
])

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Create an imputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Create a pipeline that first imputes and then fits the KNN model
knn = Pipeline([
    ('imputer', imputer),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])

knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb.fit(x_train, y_train)
y_pred_xgb = xgb.predict(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Create an imputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Create a pipeline that first imputes and then fits the SVC model
svm = Pipeline([
    ('imputer', imputer),
    ('svm', SVC())
])

svm.fit(x_train, y_train)
y_pred_svm = svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

def evaluate(model_name, y_test, y_pred):
    print(f"----- {model_name} -----")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

evaluate("Random Forest", y_test, y_pred_rf)
evaluate("KNN", y_test, y_pred_knn)
evaluate("XGBoost", y_test, y_pred_xgb)
evaluate("SVM", y_test, y_pred_svm)
```

```python
#ROC curve for XGBoost
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get predicted probabilities for class 1
y_probs = xgb.predict_proba(x_test)[:, 1]

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = roc_auc_score(y_test, y_probs)

# Plot ROC Curve
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f'XGBoost (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Random guess line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
#Hyperparameter tuning for XGBoost
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.05]
}

grid = GridSearchCV(
    estimator=XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    param_grid=param_grid,
    scoring='f1',  # you can also try 'accuracy' or 'roc_auc'
    cv=3,
    n_jobs=-1
)

grid.fit(x_train, y_train)

print("Best Hyperparameters:", grid.best_params_)
best_model = grid.best_estimator_
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred_best = best_model.predict(x_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
print("\nClassification Report:\n", classification_report(y_test, y_pred_best))
```

```python
# --- Combined Flask API and ngrok Launcher (Updated) ---

# 1. Install necessary packages (using the official pyngrok library)
!pip install flask flask-cors pyngrok

# 2. Import all required libraries
import pandas as pd
import joblib
from flask import Flask, request, jsonify
from flask_cors import CORS
from sklearn.preprocessing import LabelEncoder
import numpy as np
from pyngrok import ngrok
import threading
import os
import socket

# --- 3. Set up ngrok authentication (optional but recommended) ---
# You can get an authtoken from your ngrok dashboard: https://dashboard.ngrok.com/get-started/your-authtoken
# Replace "YOUR_NGROK_AUTHTOKEN" with your actual authtoken
NGROK_AUTHTOKEN = "30DG3yHJor3XXhBceum11ReVOrd_6hh5pksySedGe5nZS3HZr"
ngrok.set_auth_token(NGROK_AUTHTOKEN)

# --- 4. Load your trained model and prepare the data columns ---
try:
    model = joblib.load('xgboost_income_model.pkl')
    print("Model loaded successfully.")
except FileNotFoundError:
    print("Error: 'xgboost_income_model.pkl' not found. Please upload the model file.")
    model = None

try:
    # Load and clean the original dataset to get the column structure
    original_data = pd.read_csv('adult 3.csv')
    original_data.columns = [
        'age', 'workclass', 'fnlwgt', 'education', 'educational-num',
        'marital-status', 'occupation', 'relationship', 'race', 'gender',
        'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'
    ]
    original_data.replace(' ?', np.nan, inplace=True)
    original_data.dropna(inplace=True)

    le = LabelEncoder()
    original_data['gender'] = le.fit_transform(original_data['gender'])

    X_original = original_data.drop(columns=['income', 'race'])
    X_original = pd.get_dummies(X_original, drop_first=True)

    MODEL_COLUMNS = X_original.columns
    print("Model columns prepared.")

except FileNotFoundError:
    print("Error: 'adult 3.csv' not found. Please upload the dataset.")
    MODEL_COLUMNS = []

# --- 5. Initialize the Flask App ---
app = Flask(__name__)
CORS(app)

# --- 6. Define the Prediction Endpoint ---
@app.route('/predict', methods=['POST'])
def predict():
    if not model or not MODEL_COLUMNS.any():
        return jsonify({'error': 'Model or data columns not loaded properly'}), 500

    try:
        json_data = request.get_json()
        new_data = pd.DataFrame(json_data, index=[0])

        # Preprocess the new data
        new_data['gender'] = le.transform(new_data['gender'])
        new_data.drop('race', axis=1, inplace=True)
        new_data_encoded = pd.get_dummies(new_data)
        final_data = new_data_encoded.reindex(columns=MODEL_COLUMNS, fill_value=0)

        # Make prediction
        prediction = model.predict(final_data)
        prediction_proba = model.predict_proba(final_data)

        result = int(prediction[0])
        confidence = float(prediction_proba[0][result])

        return jsonify({
            'prediction': result,
            'confidence': confidence
        })

    except Exception as e:
        return jsonify({'error': str(e)}), 400

# --- 7. Start ngrok and run the App ---
def find_free_port():
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.bind(('localhost', 0))
    port = sock.getsockname()[1]
    sock.close()
    return port

port = find_free_port()

def run_app():
  # The use_reloader=False is important for stable operation in Colab
  app.run(port=port, use_reloader=False)

# Start the Flask app in a background thread
threading.Thread(target=run_app).start()

# Open a tunnel to the Flask app
public_url = ngrok.connect(port)
print(f" * Your public URL is: {public_url}")
print(" * API is now running! Copy the URL above for your website.")
```
